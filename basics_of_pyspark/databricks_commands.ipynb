```sql
%sql
CREATE DATABASE IF NOT EXISTS ecommerce;
USE ecommerce;
```

```sql
%sql
CREATE TABLE orders (
  order_id INT,
  customer_id INT,
  order_date DATE,
  order_status STRING,
  product STRING,
  quantity INT,
  price DOUBLE,
  payment_mode STRING,
  region STRING
)
USING DELTA
```

```sql
%sql
SELECT * From orders;
```

```sql
%sql
SELECT order_id, customer_id, price 
FROM orders
WHERE order_status = 'DELIVERED'; 
```

```sql
%sql
SELECT *,
    quantity * price AS total_amount
FROM orders;
```

```sql
%sql
SELECT customer_id, 
   SUM(quantity * price) AS total_spent
FROM orders
GROUP BY customer_id;
```
