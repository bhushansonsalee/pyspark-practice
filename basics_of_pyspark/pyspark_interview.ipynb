```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.window import Window

spark = SparkSession.builder.getOrCreate()

data = [
    (101, "Amit",   "IT",      60000, "2022-01-10"),
    (101, "Amit",   "IT",      70000, "2023-01-10"),
    (102, "Neha",   "HR",      45000, "2022-03-15"),
    (103, "Rahul",  "IT",      80000, "2023-01-10"),
    (104, "Pooja",  "Finance", 50000, "2021-07-20"),
    (105, "Rohit",  "HR",      60000, "2023-05-01")
]

cols = ["emp_id", "emp_name", "dept", "salary", "effective_date"]

df = spark.createDataFrame(data, cols) \
          .withColumn("effective_date", to_date("effective_date"))

df.show()
```

```python
w_dept_salary = Window.partitionBy("dept").orderBy(col("salary").desc())
w_emp_date = Window.partitionBy("emp_id").orderBy("effective_date")
```

```python
df_rownum = (
    df.withColumn("row_num", row_number().over(w_dept_salary))
      .filter(col("row_num") == 1)
      .drop("row_num")
)

df_rownum.show()
```
