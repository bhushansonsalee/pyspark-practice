```python
nums = [10, 20, 5, 20]

first = second = float("-inf")

for x in nums:
    if x > first:
        second = first
        first = x
    elif first > x > second:
        second = x

print(second)
```

```python
s = "madam"

is_palindrome = True

l, r = 0, len(s) - 1

while l < r:
    if s[l] != s[r]:
        is_palindrome = False
        break
    l += 1
    r -= 1

print(is_palindrome)
```

```python
a = [1, 3, 5]
b = [2, 4, 6]

i = j = 0
merged = []

while i < len(a) and j < len(b):
    if a[i] < b[i]:
        merged.append(a[i])
        i += 1

    else:
        merged.append(b[j])
        j += 1

merged.extend(a[i:])
merged.extend(b[j:])

print(merged)
```

```python
s = "aabbcdd"

freq = {}
for ch in s:
    freq[ch] = freq.get(ch, 0) + 1

result = None
for ch in s:
    if freq[ch] == 1:
        result = ch
        break

print(result)
```

```python
nums = [1, 2, 3, 2, 1]

seen = set()
dups = set()

for x in nums:
    if x in seen:
        dups.add(x)
    else:
        seen.add(x)

print(list(dups))
```

```python
import json

path = "/dbfs/tmp/nested.json"

data = [
    {
        "id": 1,
        "user": {"name": "Amit", "age": 30},
        "skills": ["Python", "Spark"]
    },
    {
        "id": 2,
        "user": {"name": "Neha", "age": 28},
        "skills": ["SQL", "Databricks"]
    }
]

with open(path, "w") as f:
    for record in data:
        f.write(json.dumps(record) + "\n")

print("JSON written to /dbfs/tmp")
```
