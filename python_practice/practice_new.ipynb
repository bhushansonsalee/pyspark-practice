```python
nums = [1, 2, 3, 4]
l, r = 0, len(nums) - 1
while l < r:
    nums[l], nums[r] = nums[r], nums[l]
    l += 1
    r -= 1

print(nums)
```

```python
s = "madam"

l, r = 0, len(s) - 1
is_palindrome = True

while l < r:
    if s[l] != s[r]:
        is_palindrome = False
        break
    l += 1
    r -= 1

print(is_palindrome)
```

```python
nums = [1, 2, 2, 3, 1]

seen = set()
result = []

for x in nums:
    if x not in seen:
        seen.add(x)
        result.append(x)

print(result)
```

```python
a = [1, 3, 5]
b = [2, 4, 6]

i = j = 0
merged = []

while i < len(a) and j < len(b):
    if a[i] < b[j]:
        merged.append(a[i])
        i += 1
    else:
        merged.append(b[j])
        j += 1

merged.extend(a[i:])
merged.extend(b[j:])

print(merged)
```

```python
nums = [0, 1, 0, 3, 12]

result = []
zeros = 0

for x in nums:
    if x == 0:
        zeros += 1
    else:
        result.append(x)

result.extend([0] * zeros)
print(result)
```

```python
nums = [2, 1, 5, 1, 3, 2]
k = 3

window_sum = sum(nums[:k])
max_sum = window_sum

for i in range(k, len(nums)):
    window_sum += nums[i] - nums[i - k]
    max_sum = max(max_sum, window_sum)

print(max_sum)
```
