```python
from pyspark.sql.functions import *

data = [
    (101, "2022-04-01", "Laptop", "Electronics", 70000, 10, 2, "North", "UPI", 4.5),
    (102, "2022-04-02", "Phone", "Electronics", 30000, 5, 1, "West", "Card", 4.2),
    (103, "2022-04-03", "Shoes", "Fashion", 5000, 20, 3, "South", "Cash", 4.0),
    (104, "2022-04-04", "Watch", "Fashion", 8000, 15, 2, "East", "UPI", 3.8),
    (105, "2022-04-05", "TV", "Electronics", 45000, 8, 1, "North", "Card", 4.6)
]

cols = [
    "order_id",
    "order_date", 
    "product_name",
    "category",
    "price",
    "discount_percent",
    "quantity",
    "region",
    "payment_method",
    "rating"
]

df = spark.createDataFrame(data, cols)
```

ADD COLUMNS

```python
df1 = df.withColumn(
    "discount_amount",
    col("price") * col("discount_percent") / 100
)

df1.display()
```

```python
df1 = df1.withColumn(
    "final_price",
    col("price") - col("discount_amount")
)

df1.display()
```

```python
df1 = df1.withColumn(
    "total_revenue",
    col("final_price") * col("quantity")
)
df1.display()
```

```python
df1 = df1.withColumn("country", lit("India"))
df1.display()
```

```python
df1 = df1.withColumn(
    "rating_category",
    when(col("rating") >= 4.5, "Excellent")
    .when(col("rating") >= 4, "Good")
    .otherwise("Average")
)
df1.display()
```

```python
df2 = df1.withColumn("price", col("price") * 1.05)
```

RENAME COLUMNS

```python
df3 = df2.withColumnRenamed("price", "updated_price")
```

```python
df3 = df3.withColumnRenamed("region", "customer_region") \
    .withColumnRenamed("quantity", "quantity_sold")
```

```python
df3 = df2.select(
    col("order_id"),
    col("product_name").alias("product_naame"),
    col("category"),
    col("updated_price"),
    col("discount_percent"),
    col("quantity_sold"),
    col("customer_region")
)
```
