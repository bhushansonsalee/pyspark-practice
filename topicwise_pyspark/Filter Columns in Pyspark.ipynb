```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

data = [
    (1, "Amit", "IT", 70000, 4.5),
    (2, "Neha", "HR", 45000, 4.0),
    (3, "Rahul", "IT", 80000, 4.7),
    (4, "Pooja", "Finance", 50000, 3.8),
    (5, "Rohan", "IT", None, 3.5)
]

cols = ["id", "name", "dept", "salary", "rating"]

df = spark.createDataFrame(data, cols)
df.show()
```

```python
df.filter(col("salary") > 50000).show()
```

Multiple Conditions 

```python
df.filter(
  (col("salary") > 50000) & (col("dept")  == "IT")
).show()
```

```python
df.filter(
    (col("salary") > 50000) | (col("rating") > 4.5)
).show()
```

```python
df.filter(~(col("dept") == "HR")).show()
```

```python
df.filter("salary > 50000 AND dept = 'IT'").show()
```

```python
df.filter(col("dept").isin("IT", "HR")).show()
```

```python
df.filter(col("salary").isNull()).show()
```

```python
df.filter(col("salary").isNotNull()).show()
```

```python
df.filter(col("salary").between(40000, 80000)).show()
```

```python
df.filter(col("name").like("A%")).show()
```

```python
df.filter(
    (col("salary").isNotNull()) &
    (col("salary") > 50000) & 
    (col("dept") == "IT")
).show()
```
