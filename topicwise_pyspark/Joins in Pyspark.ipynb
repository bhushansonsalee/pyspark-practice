```python
from pyspark.sql.functions import * 

orders_data = [
    (101, 1, 20000),
    (102, 2, 15000),
    (103, 3, 30000),
    (104, 4, 25000),
    (105, 6, 18000)
]

orders_cols = ["order_id", "customer_id", "amount"]

orders_df = spark.createDataFrame(orders_data, orders_cols)
orders_df.show()
```

```python
customers_data = [
    (1, "Amit", "Delhi"),
    (2, "Neha", "Mumbai"),
    (3, "Rahul", "Pune"),
    (4, "Pooja", "Bangalore"),
    (5, "Rohan", "Chennai")
]

customers_cols = ["customer_id", "customer_name", "city"]

customers_df = spark.createDataFrame(customers_data, customers_cols)
customers_df.show()
```

```python
orders_df.join(customers_df, "customer_id", "inner").show()
```

```python
orders_df.join(customers_df, "customer_id", "left").show()
```
